## Compare one year (pick the latest year) using bar chart, pie chart, and boxplot

## You must compare all regions to see how they differ in the indicator
### Compare one year (pick the latest year) using bar chart, pie chart

Note one: For population, you compare total, for the other three indicators, you compare the average.

Note two: you need to use groupby method to get the sum or mean of a column for a group (here the group is region)

```python
# Population indicator with Region column
df_region = df.query("indicator == 'SP.POP.TOTL'")
df_region = df_region[["Year", "value","Region"]]
df_region = df_region.query("Year == 2021")
df_region
```

```python
df_region.Region.unique()
```

groupby method to get the sum

```python
df_new = df_region.groupby(['Region']).agg(["sum"])
df_new
```

```python
# joining columns of multiindex to single index
df_new.columns = ['_'.join(col) for col in df_new.columns.values]
df_new.reset_index(inplace=True)
```

```python
# Singlre index columns
df_new.columns
```

## Compare one year (pick the latest year) using bar chart, pie chart
### Bar chart for 2021 Year with aggregate Sum for the regions

```python
# bar plotly chart
fig = px.bar(
    df_new, 
    x = "Region",
    y = "value_sum", 
    title = "2021 Regions", 
    color = "Region",
    template="plotly_dark"
)
fig.update_layout(showlegend=True)
fig.show()
```

### Pie chart for 2021 Year with aggregate Sum for the regions

```python
# Pie plotly chart
fig = px.pie(
    df_new, 
    names = "Region",
    values = "value_sum", 
    title = "2021 Regions", 
    color = "Region",
    template="plotly_dark"
)
fig.update_traces(textposition='inside', textinfo='label+percent+value')
fig.update_layout(showlegend=True)
fig.show()
```

## Use all countries, and pick year 2020, make a scatter plot with x being GDP per Capita and y being Life Expentancy. This plot will tell you the relationship between wealth and health.

```python
# Health is SP.DYN.LE00.IN Life expectancy and wealth is NY.GDP.PCAP.PP.CD GDP per capita
my_indicators = ["SP.DYN.LE00.IN", "NY.GDP.PCAP.PP.CD"]

df5 = df[df["indicator"].isin(my_indicators)]
df5.shape
```

```python
# Data frame with five columns "Year",	"value",	"indicator",	"Country Code",	"Country Name"
df5 = df5[["Year",	"value",	"indicator",	"Country Code",	"Country Name"]]
df5
```

```python
# Pivot table to get seperate life expectancy and GDP per capita columns
df6 = pd.pivot_table(df5, values= 'value', index=['Country Code',	'Country Name', 'Year'], columns = 'indicator').reset_index()
df6.head()
```

```python
# Drop null values 
df6.dropna(inplace = True)
df6.head()
```

## Relationship between wealth and health using Scatter plot

```python
# Scatter plotly chart
fig = px.scatter(
    df6.query("Year == 2020"), 
    x= "NY.GDP.PCAP.PP.CD",
    y = "SP.DYN.LE00.IN", 
    title = "Wealth vs Health", 
    color="Country Name",
    template = 'plotly_dark'
)
fig.update_layout(showlegend=True)
fig.show()
```
