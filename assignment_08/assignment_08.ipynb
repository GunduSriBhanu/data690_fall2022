## Use Pandas to read the file census_costs.txt file

```python
#import libraries

import pandas as pd
import plotly.express as px
import re
```

```python
#open file and stored as list

with open("/content/census_costs.txt", "r") as f:
    lines = f.readlines()

lines
```

```python
#displaying columns

census_columns = lines[1]
census_columns
```

```python
#displaying data in list

census = lines[2:]
census
```

```python
#cleansing of columns in a list

census_columns = re.split(r'\t+', census_columns)
census_columns
```

```python
#converting list to dataframe

df = pd.DataFrame(census).rename(columns = {0:'data'})
df.head()
```

```python
# Assigning columns from cleansing into data frame

df[['Census Year', 'Total Population', 'Census Cost', 'Average Cost Per Person\n']] = df.data.str.split("\t", expand=True)
df.head()
```

```python
#dropping columns and chaniging name of column

df = df.drop('data',axis=1)
df = df.rename(columns = {'Average Cost Per Person\n' : 'Average Cost Per Person'})
```

## Data Cleansing

## Cleanse the data to make the year, population, cost, cost per person numerical type

Code functions to cleanse the columns. One function for one column
use apply() method of data frame to apply the functions to the columns for data cleansing

```python
#reeplacing special characters and unwanted strings from columns after analysis

df["Census Year"] = df["Census Year"].apply(lambda x:x.replace('*',''))

df["Census Cost"] = df["Census Cost"].apply(lambda x:x.replace('$',''))

#df["Average Cost Per Person"] = df["Average Cost Per Person"].apply(lambda x:x.replace(r'$', '').replace(r'cents', ''))

```

 Converting the "Average Cost Per Person" column to dollars with float data type  and cleaning the column

```python
df["Average Cost Per Person"] = df["Average Cost Per Person"].apply(lambda x:x.replace(r'cents', ''))

df["Average Cost Per Person"] = df["Average Cost Per Person"].apply(lambda x:x.replace(r'$', ''))

# Changing datatype for column "Average Cost Per Person"
df["Average Cost Per Person"] = df["Average Cost Per Person"].astype('float')

# Converting only values which are in cents
df["Average Cost Per Person"][:-5] = df["Average Cost Per Person"][:-5]*0.01

df
```

```python
#df["Census Year"] = df["Census Year"].str.replace(r'*', '', regex=True)
#df["Census Cost"] = df["Census Cost"].str.replace(r'$', '', regex=True)
#df["Average Cost Per Person"] = df["Average Cost Per Person"].str.replace(r'$', '', regex=True).str.replace(r'\n', '', regex=True).str.replace(r'cents', '', regex=True)
```

```python
#Removing multiple commas from the column

lst = []
for row in df['Total Population']:
  lst.append(re.sub(",","",row))
df['Total Population'] = lst
```

```python
#Removing multiple commas from the column

lst1 = []
for row in df['Census Cost']:
  lst1.append(re.sub(",","",row))
df['Census Cost'] = lst1
#df["Census Cost"] = df["Census Cost"].str.replace(r'Billion', '', regex=True).astype('float')

# removing "Billion" and changing data type
df["Census Cost"] = df["Census Cost"].apply(lambda x:x.replace(r'Billion', '')).astype('float')
```

```python
#converting Billion to value
df["Census Cost"][-2:] = df["Census Cost"][-2:]*1000000000
```

### data to make the year, population, cost, cost per person numerical type

```python
# Changing datatype for columns

df[['Census Year',	'Total Population']] = df[['Census Year',	'Total Population']].astype('int')
#df['Average Cost Per Person'] = df['Average Cost Per Person'].astype('float')
```

## Use Plotly to plot the average cost per person over time.

```python
# Plotly bar chart in ascending order
fig = px.bar(
    df, 
    x = "Census Year",
    y = "Average Cost Per Person", 
    title = "Cost per person over time", 
    color = "Census Year",
    template="plotly_dark"
)
fig.update_layout(showlegend=True)

fig.show()
```

## Use Plotly to plot the scatter plot of Total Population vs Total Cost.

```python
# Scatter plotly chart
fig = px.scatter(
    df, 
    x= "Total Population",
    y = "Census Cost", 
    title = "Population Vs Cost", 
    color="Total Population",
    template = 'plotly_dark'
)
fig.update_layout(showlegend=True)
fig.show()
```
