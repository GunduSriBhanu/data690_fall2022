### Analysing the data for the year 2013

For the year 2013, analysing the population for top 10 countries with highest population with bar graph. Among these China, India and United states are the highest populated coutries.

```python
df2 = df.query("indicator == 'SP.POP.TOTL' and Year == 2013") 
# Bar Plotly chart for top 10 populated countries
fig = px.bar(
    df2.sort_values(by="value", ascending=False)[0:10], 
    x="Country Name",
    y = "value", 
    title = "2013 Population of G20 Countries", 
    color = "Country Name", 
    template="plotly_dark"
)
fig.update_layout(showlegend=True)
fig.show()
```

Further analysing the data with population for G20 countries whose age is more than 25 years.

```python
#bar plot for top 10 populated countries with age 25+
fig = px.bar(
    df.query("indicator == 'SP.POP.AG25.TO.UN' and Year == 2013").sort_values(by="value", ascending=False)[0:10], 
    x="Country Name",
    y = "value", 
    title = "25+ Population of G20 Countries in 2013", 
    color = "Country Name", 
    template="plotly_dark"
)
fig.update_layout(showlegend=True)
fig.show()
```

Among these, China and India population are almost equal with population of more than 25 years of age followed by Unites States.

## Analysis based on Regions

```python
# Analysis based on regions with mean aggregate
df_unemp = df.query("indicator == 'SL.UEM.TOTL.ZS'")
df_unemp = df_unemp[["Year", "value","Region"]]
df_unemp = df_unemp.groupby(['Year','Region']).agg(["mean"])
df_unemp.columns = ['_'.join(col) for col in df_unemp.columns.values]
df_unemp.reset_index(inplace=True)
df_unemp[0:5]
```

```python
# Bar plot with mean aggregate
fig = px.bar(
    df_unemp.query('Year == 2013').sort_values(by="value_mean", ascending=False),  
    x = "Region",
    y = "value_mean", 
    title = "2013 Regions of unemployment", 
    color = "Region",
    template="plotly_dark"
)
fig.update_layout(showlegend=True)
fig.show()
```

### From my observation,
**Sub-Saharan Africa** region has highest unemployment in the year 2013

### Comparison of unemployment with expenditure on education

```python
# Filtering data of unemployment and expenditure indicators
my_indicators = ["SL.UEM.TOTL.ZS", "NY.ADJ.AEDU.GN.ZS"]

df_unemp_expend = df[df["indicator"].isin(my_indicators)]
df_unemp_expend.shape
```

```python
# Data with the selected columns
df_unemp_expend = df_unemp_expend[["Year",	"value",	"indicator",	"Country Code",	"Country Name"]]
df_unemp_expend
```

```python
# pivot table for the indicators of unemployment and expenditure
df_eem_exp = pd.pivot_table(df_unemp_expend, values= 'value', index=['Country Code',	'Country Name', 'Year'], columns = 'indicator').reset_index()
df_eem_exp.head()
```

```python
# Scatter plotly chart for unemployment variation with expenditure in year 2013
fig = px.scatter(
    df_eem_exp.query("Year == 2013"), 
    x= "NY.ADJ.AEDU.GN.ZS",
    y = "SL.UEM.TOTL.ZS", 
    title = "Unemployment vs Expenditure on Education", 
    color="Country Name",
    template = 'plotly_dark'
)
fig.update_layout(showlegend=True)
fig.show()
```

### From the observation,
- **Korea and Japan** with ***lowest*** unemployment rate with medium contribution on education.
 
- **South Africa** with ***highest*** unemployment rate eventhough the expenditure on education is highes t for South Africa.

### Expenditure of Education over years for the countries Australia, China, India, United States, and United Kingdom.

```python
# Comparison with aggregate mean
df_10_15 = df_sc.query("indicator == 'NY.ADJ.AEDU.GN.ZS'")
df_10_15 = df_10_15[["Year", "value",'Country Code']]
df_10_15 = df_10_15.groupby(['Year','Country Code']).agg(["mean"])
df_10_15.columns = ['_'.join(col) for col in df_10_15.columns.values]
df_10_15.reset_index(inplace=True)
df_10_15[0:5]
```

```python
# Comparison of line chart for all the years 
fig = px.line(
    df_10_15, 
    x="Year",
    y = "value_mean", 
    title = "expenditure on education over Years", 
    color='Country Code',
    template = 'plotly_dark'
)
fig.update_layout(showlegend=True)
fig.show()
```

### Unemployment over years for the countries Australia, China, India, United States, and United Kingdom.

```python
# Unemployment comparison over years with aggregate mean
df_10_15_unemp = df_sc.query("indicator == 'SL.UEM.TOTL.ZS'")
df_10_15_unemp = df_10_15_unemp[["Year", "value",'Country Code']]
df_10_15_unemp = df_10_15_unemp.groupby(['Year','Country Code']).agg(["mean"])
df_10_15_unemp.columns = ['_'.join(col) for col in df_10_15_unemp.columns.values]
df_10_15_unemp.reset_index(inplace=True)
df_10_15_unemp
```

```python
# Comparison of line chart for all the years 2010-2015
fig = px.line(
    df_10_15_unemp, 
    x="Year",
    y = "value_mean", 
    title = "Unemployment over Years", 
    color='Country Code',
    template = 'plotly_dark'
)
fig.update_layout(showlegend=True)
fig.show()
```
